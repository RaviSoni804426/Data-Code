```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3 = np.percentile(col,[25,75])
    IQR=Q3-Q1
    upper_bound=Q3+1.5*IQR
    lower_bound=Q3-1.5*IQR
    return upper_bound,lower_bound
```
