```python
import torch
import torch.nn.functional as F
```

```python
image = torch.tensor([
    [1,2],
    [3,4]]).float().unsqueeze(0).unsqueeze(0)
```

```python
# Nearerest-neighbour upsampling
upsamppled_nn = F.interpolate(image, scale_factor=2, mode='nearest')
```

```python
# Transposed COnvolution
conv_transpose = torch.nn.ConvTranspose2d(1, 1, kernel_size=3,stride=3, padding=1)
output_transpose = conv_transpose(image)
```

```python
print("Nearest neighbour Upsampling : ", upsamppled_nn.squeeze())
print("Transpose Convolution output : ", output_transpose.squeeze())
```
